```python
from quixbugs import QuixBugsDataset

from rich import print

import pandas as pd

quix_bugs_dataset = QuixBugsDataset("python")
prog_info = quix_bugs_dataset.prog_info

df = pd.DataFrame(prog_info)

df
```

```python
a = df.groupby(["bug_type"]).size().to_dict()
b = df.groupby(["bug_consequence"]).size().to_dict()
c = df.groupby(["bug_impact"]).size().to_dict()

print(a,b,c)
```

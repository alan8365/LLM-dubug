```python
from glob import glob
from pathlib import Path
import re

generated_code_path = Path('generated_code') / 'gpt35'

all_file = generated_code_path.glob('**/*.py')
all_file = [x for x in all_file if re.match('.*_\d\.py', str(x))]
all_file = sorted(all_file, key=lambda x: x.name)

print(all_file)
```

```python
for index in range(len(all_file)):
    file = all_file[index]

    with open(file, 'r') as f:
        content = f.read()

    print(content)
```
